```python
#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'langkahTercepat' function below.
#
# The function is expected to return an INTEGER.
# The function accepts following parameters:
#  1. 2D_INTEGER_ARRAY ladders
#  2. 2D_INTEGER_ARRAY snakes
#

def langkahTercepat(ladders, snakes):
    # Create a dictionary to represent the game board
    board = {i: i for i in range(1, 101)}

    # Update the board for ladders and snakes
    for start, end in ladders:
        board[start] = end
    for start, end in snakes:
        board[start] = end
        
    visited = set()
    queue = [(1, 0)]

    while queue:
        position, moves = queue.pop(0)

        if position == 100:
            break

        for i in range(1, 7):
            if (position + i) > 100:
                next_position = board[position + i - 100]
            else:
                next_position = board[position + i]

            if next_position not in visited:
                visited.add(next_position)
                queue.append((next_position, moves + 1))
    return moves

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    L = int(input().strip())

    ladders = []

    for _ in range(L):
        ladders.append(list(map(int, input().rstrip().split())))

    S = int(input().strip())

    snakes = []

    for _ in range(S):
        snakes.append(list(map(int, input().rstrip().split())))

    answer = langkahTercepat(ladders, snakes)

    fptr.write(str(answer) + '\n')

    fptr.close()
```

```python
#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'calculateStamp' function below.
#
# The function is expected to return an INTEGER.
# The function accepts following parameters:
#  1. INTEGER N
#  2. INTEGER M
#  3. 2D_STRING_ARRAY paper
#

# Check the biggest square can created by an '_'
def can_stamp(i,j,s, mtrx):
    nrow, ncol = len(mtrx), len(mtrx[0])
    
    if i+s > nrow or j+s > ncol:
        return False
    
    for x in range(i, i+s):
        for y in range(j, j+s):
            if mtrx[x][y] == "#":
                return False
    return True

# Check if stamp size s can cover all in mtrx
def can_cover_all(s, mtrx):
    nrow, ncol = len(mtrx), len(mtrx[0])
    
    visited = [[False for _ in range(ncol)] for _ in range(nrow)]
    for i in range(nrow - s + 1):
        for j in range(ncol - s + 1):
            if can_stamp(i, j, s,mtrx):
                for x in range(i, i + s):
                    for y in range(j, j + s):
                        visited[x][y] = True
    for i in range(nrow):
        for j in range(ncol):
            if mtrx[i][j] == "_" and not visited[i][j]:
                return False
    return True

def calculateStamp(N, M, paper):
    
    # Fix the input format
    mtrx = [lx[0] for lx in paper]
    
    #Handle special cases
    if not(mtrx):
        return 0
    elif len(mtrx) == 1:
        return 1
    
    #Check all biggest square can created by all '_'
    nrow, ncol = len(mtrx), len(mtrx[0])
    
    max_possible_size = min(nrow, ncol)
    for size in range(max_possible_size, 0, -1):
        if can_cover_all(size, mtrx):
            return size
    return 0
    
    

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    first_multiple_input = input().rstrip().split()

    N = int(first_multiple_input[0])

    M = int(first_multiple_input[1])

    paper = []

    for _ in range(N):
        paper.append(input().rstrip().split())

    result = calculateStamp(N, M, paper)

    fptr.write(str(result) + '\n')

    fptr.close()
```
