```python
def get_common_letters(s1,s2):
    s1_set = set(s1)
    s2_set = set(s2)
    intersect_set = s1_set.intersection(s2_set)
    return intersect_set

def is_substring_formable(substr, parent_str):
    i = 0
    for c in parent_str:
        if i >= len(substr):
            return True
        if c == substr[i]:
            i+= 1
    return i >= len(substr)

def get_all_remove_poss(strx, n, parent):
    if is_substring_formable(strx, parent):
        return len(strx)
    elif n == 0:
        return 0
    
    poss_lst = []
    for i in range(0,len(strx)):
        poss_x = get_all_remove_poss(strx[:i] + strx[i+1:],n-1, parent)
        if poss_x:
            return poss_x
    return poss_lst
        
def lcs(s1,s2):
    m = len(s1)
    n = len(s2)
    
    dp = [[0] * (n+1) for x in range(m+1)]
    
    for i in range(1,m+1):
        for j in range(1,n+1):
            if s1[i-1] == s2[j-1]:
                dp[i][j] = dp[i-1][j-1] + 1
            else:
                dp[i][j] = max(dp[i-1][j],dp[i][j-1])
    return dp[m][n]

        
# def lcs(s1,s2,m,n,memo):
#     if m == 0 or n == 0:
#         return 0
    
#     if memo[m][n] != -1:
#         return memo[m][n]
    
#     if s1[m-1] == s2[n-1]:
#         memo[m][n] = 1+ lcs(s1,s2,m-1,n-1, memo)
#         return memo[m][n]
    
#     memo[m][n] = max(lcs(s1,s2,m,n-1,memo), lcs(s1,s2,m-1,n-1, memo))
#     return memo[m][n]
        
    
def commonChild(s1, s2):
    # Write your code here
    intersect_set = get_common_letters(s1,s2)
    print(intersect_set)
    common_s1 = ''.join([c for c in s1 if c in intersect_set])
    common_s2 = ''.join([c for c in s2 if c in intersect_set])
    
    m = len(common_s1)
    n = len(common_s2)
    
    memo = [[-1 for _ in range(n + 1)] for _ in range(m + 1)]
    return lcs(common_s1,common_s2)
        
    
    
```
