```python
def morganAndString(str_a, str_b):
    len_a, len_b = len(str_a), len(str_b)
    idx_a, idx_b = 0, 0
    result = []

    while idx_a < len_a and idx_b < len_b:
        if str_a[idx_a:] < str_b[idx_b:]:
            result.append(str_a[idx_a])
            idx_a += 1
        else:
            result.append(str_b[idx_b])
            idx_b += 1

    result += str_a[idx_a:]
    result += str_b[idx_b:]

    return ''.join(result)
```
