```python
def lcs(s1,s2):
    m,n = len(s1), len(s2)
    dp = [[0] * (n+1) for x in range(0,m+1)]
    for i in range(0,m+1):
        for j in range(0,n+1):
            if s1[i-1] == s2[j-1]:
                dp[i][j] = dp[i-1][j-1] + 1
            else:
                dp[i][j] = max(dp[i][j-1], dp[i-1][j])
    return dp[m][n]
```
